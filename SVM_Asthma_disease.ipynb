```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
# read csv file using pandas and convert in into dataframe
df = pd.read_csv('./CSV_Files/processed-data.csv')
print("Shape-> ",df.shape)
df.head(10)
```

```python
Non_dupDF = df.drop_duplicates()
Non_dupDF
```

```python
print(Non_dupDF.info())
Non_dupDF.describe()
```

```python
Non_dupDF.isnull().sum()
```

```python
Non_dupDF.rename(columns={'Severity_None': 'Target'}, inplace=True)
Non_dupDF.head()
```

```python
### 0 = male and 1 = female


X = Non_dupDF.drop(['Severity_Mild', 'Severity_Moderate','Gender_Male','Target'], axis='columns')
y = Non_dupDF.Target
print(X.shape, y.shape)
```

```python
corr_matrix = Non_dupDF.corr()
corr_matrix['Target']*100
```

```python
### correlation matrix in heatmap
plt.figure(figsize=(20,7))
sns.heatmap(corr_matrix, annot=True, cmap='RdYlGn')
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)

print(len(X_train), len(X_test))
print(len(X_train)+len(X_test))
```

```python
from sklearn.svm import SVC
model = SVC()
```

```python
model.fit(X_train,y_train)
```

```python
train_acc = model.score(X_train,y_train)*100
test_acc = model.score(X_test,y_test)*100
print(train_acc, test_acc)
```

```python
y_pred = model.predict(X_test)
from sklearn.metrics import confusion_matrix, classification_report, mean_absolute_error, mean_squared_error
cm = confusion_matrix(y_test, y_pred)
print(cm)
print('classification_report', classification_report(y_test, y_pred))

### visualization of cm
plt.figure(figsize=(7,3))
sns.heatmap(cm, annot=True)
```

```python
### Errors in this model
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)


print('mae => ', mae)
print('mse => ', mse)
print('rmse => ', rmse)
```
