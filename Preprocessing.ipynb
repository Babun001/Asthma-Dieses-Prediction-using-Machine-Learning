```python
# importing all needed libraries..
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# read csv file using pandas and convert in into dataframe
df = pd.read_csv('./CSV_Files/processed-data.csv')
print("Shape-> ",df.shape)
df.head(10)
```

```python
# checking for null values
print(df.isnull().sum())
```

```python
print(df.describe())
df.shape
```

```python
Non_dupDF = df.drop_duplicates()
Non_dupDF
```

```python
Non_dupDF.rename(columns={'Severity_None': 'Target'}, inplace=True)
# Non_dupDF.Target[Non_dupDF.Target==0].count() ### 1 -> 1440  0-> 4320
```

```python
print(Non_dupDF.shape)
# (316800 - 5760) = 311040 are duplicated values are cleared
```

```python
corr_matrix = Non_dupDF.corr()
corr_matrix
corr_matrix['Target']*100
```

```python
# it is time to check coorelation
plt.figure(figsize=(20,7))
sns.heatmap(corr_matrix,annot=True, cmap='Blues')
```
