```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
# read csv file using pandas and convert in into dataframe
df = pd.read_csv('./CSV_Files/processed-data.csv')
print("Shape-> ",df.shape)
df.head(10)
```

```python
Non_dupDF = df.drop_duplicates()
Non_dupDF
```

```python
print(Non_dupDF.info())
Non_dupDF.describe()
```

```python
Non_dupDF.isnull().sum()
```

```python
Non_dupDF.rename(columns={'Severity_None': 'Target'}, inplace=True)
Non_dupDF.head()
```

```python
### 0 = male and 1 = female


X = Non_dupDF.drop(['Severity_Mild', 'Severity_Moderate','Gender_Male','Target'], axis='columns')
y = Non_dupDF.Target
print(X.shape, y.shape)
```

```python
plt.scatter(Non_dupDF['Gender_Female'],y)
```

```python
corr_matrix = Non_dupDF.corr()
corr_matrix['Target']*100
```

```python
### correlation matrix in heatmap
plt.figure(figsize=(20,7))
sns.heatmap(corr_matrix, annot=True)
```
